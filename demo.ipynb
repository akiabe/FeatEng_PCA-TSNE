```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
print(df.shape)
```

```python
from sklearn import decomposition

X = df.drop(columns=["id", "target"], axis=1).values
y = df.target.values

pca = decomposition.PCA(n_components=2)
pca.fit(X)
X_2D_pca = pca.fit_transform(X)

print(X_2D_pca.shape)
```

```python
df.loc[:, "X_2D_pca_[0]"] = X_2D_pca[:, 0]
df.loc[:, "X_2D_pca_[1]"] = X_2D_pca[:, 1]
df.shape
```

```python
from sklearn import manifold

tsne = manifold.TSNE(n_components=2, random_state=0)
X_2D_tsne = tsne.fit_transform(X)
print(X_2D_tsne.shape)
```

```python
from sklearn import neighbors

knn = neighbors.KNeighborsClassifier(n_neighbors=2)
knn.fit(X, y)
dist, idx = knn.kneighbors(X)
```

```python
print(dist.shape)
print(idx.shape)
```
